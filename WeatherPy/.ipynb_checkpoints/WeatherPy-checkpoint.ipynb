# WeatherPy
Observations:
1) Temperature and Latitude have a strong relationship. The lower you get in latitude, the higher the temperature. In the Latitude vs temperature scatter plot, you can clearly see an almost horsehoe shape, clearly showing the further you go away from 0 or the equater, the colder the max temp will be.
2)The humidity vs latitude scatter plot showed me that there was not a strong relationsip between the two metrics. It was interesting to see that there were a lot of cities with above 60% humidity. This time of year, the northern hemisphere is transitioning to winter and the sothern hemisphere is transitioning into summer so i can see that the north could be getting some rain/snow and the southern hemisphere getting into their rainy season.
3)Out of all the data comparisons, the wind speed and latitude had the least correlation. The wind for the majority of the cities were below 10mph. The slope in the linear regression plot also revealed there was no correlation with latitude. The slopes for both the northern and southern hemisphere were below 0.19.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
temp_units = "imperial"

query_url = f"http://api.openweathermap.org/data/2.5/weather?appid={weather_api_key}&units={temp_units}&q="
```

```python
#Initiate counter for city number below
number = 0
set_number=1
#Lists for holding required items
city_name = []
lat = []
lng = []
temp = []
humid = []
clouds = []
wind = []
countries=[]
dates=[]
#Loop through each city, extract required items, and print result
print('------------------------------')
print('Beginning Data Retrieval')
print('------------------------------')

for city in cities:
    try:
        response = requests.get(query_url + city.replace(" ","&")).json()
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        temp.append(response['main']['temp'])
        humid.append(response['main']['humidity'])
        clouds.append(response['clouds']['all'])
        wind.append(response['wind']['speed'])
        countries.append(response['sys']['country'])
        dates.append(response['dt'])
    
        if number > 49:
            number = 1
            set_number += 1
            city_name.append(response['name'])
        else:
            number += 1
            city_name.append(response['name'])
        print(f'Processing Record {number} of Set {set_number} | {city}')
    except Exception:
        print("City not found.Skipping...")
        
print('------------------------------')
print('Data Retrieval Complete')
print('------------------------------')
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
city3 = {"City":(city_name),"Lat":(lat),"Lng":(lng),"Max Temp":(temp),"Humidity":(humid),"Cloudiness":(clouds),"Wind Speed":(wind),"Country":(countries),"Date":(dates)}
city_df=pd.DataFrame(city3)
city_df.head()
```

```python
#output to CSV
city_df.to_csv(output_data_file)
```

## Inspect the data and remove the cities where the humidity > 100%.
----
Skip this step if there are no cities that have humidity > 100%. 

```python
#Count of cities with humidity over 100
humid_over= city_df.loc[city_df["Humidity"] > 100].count()
humid_over
```

```python
# Make a new DataFrame equal to the city data to drop all humidity outliers by index.

clean_city_data=city_df.sort_values('Humidity',ascending=True, inplace=False)
quartiles=clean_city_data['Humidity'].quantile([.25,.50,.75])
lowerq=round(quartiles[.25],2)
upperq=round(quartiles[.75],2)
iqr=round((upperq-lowerq),2)
lower_bound = round((lowerq - (1.5*iqr)),2)
upper_bound = round((upperq + (1.5*iqr)),2)

# Passing "inplace=False" will make a copy of the city_data DataFrame, which we call "clean_city_data".
clean_city_data_hum= clean_city_data.loc[clean_city_data["Humidity"] > lower_bound]
clean_city_data_hum2=clean_city_data_hum.loc[clean_city_data_hum["Humidity"] < upper_bound]
clean_city_data_hum2

# Passing "inplace=False" will make a copy of the city_data DataFrame, which we call "clean_city_data".
```

## Linear Regression

```python
north= city_df.loc[city_df["Lat"] > 0]
south=city_df.loc[city_df["Lat"] < 0]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
north= city_df.loc[city_df["Lat"] > 0]
xvalue=north["Lat"]
yvalue=north["Max Temp"]
#Get linear regression formula
(slope,intercept,rvalue,pvalue,stderr)=linregress(xvalue,yvalue)
regress_values=xvalue*slope+intercept
line_eq= "y=" + str(round(slope,2)) + "x +" + str(round(intercept,2))
#Format scatter plot
plt.scatter(xvalue,yvalue)
plt.xlabel("Latitude")
plt.ylabel("Max Temp")
plt.title("Northern Hemisphere - Max Temp vs. Latitude Linear Regression")
plt.plot(xvalue,regress_values, "r-")
plt.annotate(line_eq,(4,-10),fontsize=20,color="red")
print(f"The rsquared is: {round(rvalue**2,2)}")
print(f"The line equation is:{(line_eq)}")
```
